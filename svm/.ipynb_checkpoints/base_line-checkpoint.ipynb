```python
from sklearn.svm import LinearSVC 
import pickle as p
import logging
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data_path = '../data/'

word_tfidf = p.load(open(data_path+'word.tfidf', 'rb'))
token_tfidf = p.load(open(data_path+'token.tfidf', 'rb'))

# because the nums of calss 4,5 noly 1,2, so just drop to 0 class
y = pd.read_csv(data_path+'call_reason.csv', usecols=['label'])['label'].apply(lambda x: x if x < 4 else 0)
```

```python
def eval(x, y, name):
        clf = LinearSVC()
        train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
        clf.fit(train_x, train_y)
        pred_y = clf.predict(test_x)

        cormat = confusion_matrix(test_y, pred_y)
        for i in range(cormat.shape[0]):
            cormat[i][i] = 0
        corrmat = StandardScaler().fit_transform(cormat)
        f, ax = plt.subplots(figsize=(10, 10))
        sns.heatmap(cormat, xticklabels = ['others', 'transform', 'research', 'complain'],\
                    yticklabels = ['others', 'transform', 'research', 'complain'],square=True, linewidths=.5, annot=True, cmap="YlGnBu")
        plt.show()
        
eval(word_tfidf, y, 'word_level')
eval(token_tfidf, y, 'token_level')
```
